```python
# Creates a seperate dataframe
df3 = (df[['age', 'cholesterol', 'resting bps', 'resting ecg', 'max heart rate', 'oldpeak', 'exercise angina', 'target']]) 
```

```python
# Groups data by a column and applies a sum aggregation.
# Gives us data to identify trends and patterns related to age using our 3rd dataframe
df3.groupby('age').sum()
```

```python
# Define a function to calculate the range of a column
def calculate_range(column):
    return column.max() - column.min()

# Apply the function to each column
range_per_column = df.apply(calculate_range)

print(range_per_column)
```

```python
# Define a function to categorize patients based on cholesterol and heart disease presence
def categorize_risk(row):
    if row['cholesterol'] > 250 and row['target'] == 1:
        return 'high_risk'
    else:
        return 'low_risk'

# Apply the function to each row and create a new column 'risk_category'
df3['risk_category'] = df3.apply(categorize_risk, axis=1)

print(df3)
```

```python
# Aggregates data and creates visualization bar chart to better understand trends
import matplotlib.pyplot as plt

# Group by 'age' and sum the values
grouped_sum = df.groupby('age').sum()

grouped_sum['cholesterol'].plot(kind='bar')
plt.title('Total Cholesterol Levels by Age Group')
plt.ylabel('Total Cholesterol')
plt.show()
```

```python
# Create a larger bar chart
plt.figure(figsize=(14, 8))  # Larger figure size
grouped_sum['cholesterol'].plot(kind='bar', color='skyblue', edgecolor='black')

# Adding titles and labels
plt.title('Total Cholesterol Levels by Age Group', fontsize=18)
plt.xlabel('Age', fontsize=16)
plt.ylabel('Total Cholesterol', fontsize=16)
plt.xticks(rotation=0, fontsize=14)
plt.yticks(fontsize=14)

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Applies encoding to dataset to compute catergorical data and make accurate predictions
from sklearn.preprocessing import LabelEncoder

from sklearn.preprocessing import LabelEncoder

# One-Hot Encoding for nominal categorical variables
df = pd.get_dummies(df, columns=['cp', 'restecg', 'slope', 'ca', 'thal'])

# Label Encoding for binary categorical variables
binary_columns = ['sex', 'fbs', 'exang', 'target']

le = LabelEncoder()
for col in binary_columns:
    df[col] = le.fit_transform(df[col])

# Display the first few rows of the encoded dataset
print(df.head())
```
